```python
### While loop - keep on getting executed till the condition is true
i = 1
while i < 5:
    print(i)
    i += 1
```

```python
## Loop Control Statements
## Break - It exits the loop prematurely 
for i in range(10):
    if i == 5 :
        break
    print(i)
```

```python
## Continue - Skips the current itertaion and continue with the rest 
for i in range(10):
    if i % 2 == 0 :
        continue
    else :
        print(i)
```

```python
### Pass - Is a null operation , and it does nothing 
for i in range(5):
    if i == 3:
        pass
    print(i)
```

```python
## Nested Loops - loop inside a loop
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j : {j}")
```
