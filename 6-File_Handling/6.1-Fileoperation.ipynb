```python
print("File Operation")
```

```python
## Read a whole file
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
## Read a file line by line 
with open("example.txt", 'r') as file:
    for line in file:
        print(line.strip()) ##Strip removes the new line character
```

```python
## Writing a file(Overwriting)
with open("example.txt", 'w') as file:
    file.write("Hello world!\n")
    file.write("This is a new line \n")
```

```python
## Write a file (without overwriting)
with open("example.txt", 'a') as file :
    file.write("Append operation is taking place. \n")
```

```python
### Writing a list of lines to a file
lines = ['This is my first line.\n',
         'This is my second line.\n',
         'This is my third line.\n']

with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
### Binary Files
## Writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
##Reading a binary file
with open('example.bin', 'rb') as file:
    content = file.read()
    print(content)
```

```python
### Read the content from a source text file and write to a destination text file
##Copying a file
with open("example.txt", 'r') as source:
    content = source.read()

with open('destination.txt', 'w') as destination:
    destination.write(content)
```

```python
## Read a text files and count the numbers of lines word and characters
with open("example.txt", 'r') as file:
    lineCount = 0
    wordCount = 0
    charCount = 0
    for line in file:
        lineCount += 1
        wordCount += len(line.split())
        charCount += len(line)
    print(f"The number of lines is : {lineCount}")
    print(f"The number of words is : {wordCount}")
    print(f"The number of characters is : {charCount}")
```

```python
## Writing and then reading a file
with open('example.txt', 'w+') as file :
    file.write("Hello world\n")
    file.write("This a new line.\n")

    ##Move the file cursor to the beginning
    file.seek(0)

    ##Read the content of the file
    print(file.read())
```
