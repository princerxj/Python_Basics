```python
## Syntax
lambda arguments: expression
```

```python
def addition(a, b) :
    return a + b
addition(2,3)
```

```python
add = lambda a,b : a + b
print(type(add))
print(add(1,7))
```

```python
def even( num ) :
    return num % 2 == 0
even(24)
```

```python
even1 = lambda num : num % 2 == 0
even1(78)
```

```python
even1(79)
```

```python
def addition(x, y, z) :
    return x + y + z
addition(12,13,14)
```

```python
addition1 = lambda x,y,z : x + y + z
addition1(9,8,18)
```

```python
### map() - It applies a function to all the items in a list 
numbers = [1,2,3,4,5,6]
def square(x):
    return x ** 2
square(2)
```

```python
list(map(lambda x:x**2, numbers))
```
