```python
## try , except, else block 
try :
    num = int(input('Enter a number : '))
    result = 10 / num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError :
    print("You cannot enter denominator equal to 0")
except Exception as ex :
    print(ex)
else :
    print(f"The result is {result}")
```

```python
## try , except, else block 
try :
    num = int(input('Enter a number : '))
    result = 10 / num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError :
    print("You cannot enter denominator equal to 0")
except Exception as ex :
    print(ex)
else : ## Gets executed only when there is no exception 
    print(f"The result is {result}")
finally: ## Gets executed anyhow no matter whether there  is an exception or not
    print("Execution Completed.")
```

```python
### File handling and exception handling 
try :
    file = open("example.txt", 'r')
    content = file.read()
    print(content)
except FileNotFoundError:
    print("The file doesnot exist")
finally :
    if 'file' in locals() or not file.closed():
        file.close()
        print("File Closed")
```
