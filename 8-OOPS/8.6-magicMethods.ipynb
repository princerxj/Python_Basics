# Magic Methods 
magic methods in Python, also known as dunder methods (double underscore methods), are special methods that start and end with double underscores. These methods enable you to define the behaviour of the objects for built-in operations, such as arithmetic ops, comparisons and many more. 

## Magic Methods 
Magic methods are the predefined methods in python that we can override to change teh behaviour of your objects . Some common magic methods include : 

```python
'''
__init__ : Initialise a new instance of a class 
__str__ : Returns the string representation of an object 
__repr__ : returns an official string representation of an object 
__len__ : returns the length of an object 
__getitem__ : Gets an item from the container 
__setitem__ : Sets an item in the container 
'''
```

```python
## Basic Methods 
class Person : 
    def __init__(self, name, age) :
        self.name = name 
        self.age = age 

person = Person("Krish", 34)
print(person)
```

```python
## Basic Methods 
class Person : 
    def __init__(self, name, age) :
        self.name = name 
        self.age = age 
    
    def __str__(self):
        return f"{self.name}, {self.age} years old"
    
    def __repr__(self) :
        return f"Person(name = {self.name}, age = {self.age})"

person = Person("Krish", 34)
print(person)
print(repr(person))
```
