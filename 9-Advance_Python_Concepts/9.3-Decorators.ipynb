```python
## Function Copy 
def welcome() :
    return "Welcome to the Advance Python Course"

welcome() 
```

```python
wel = welcome
print(wel())
del welcome
print(wel())
```

```python
## Closures (Function inside a function) 

def main_welcome(msg) :
    def sub_welcome_method() :
        print("Welcome to the advanced Python Course")
        print(msg)
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
main_welcome("Hello Everyone")
```

```python
def main_welcome(func, lst) :
    def sub_welcome_method() :
        print("Welcome to Python BootCamp")
        print(func(lst))
        print("Today my last day of learning python basics in this course")
    return sub_welcome_method()
```

```python
main_welcome(len, [1,2,3,4])
```

```python
### Decorator 
def main_welcome(func) :
    def sub_welcome_method() :
        print("Welcome to Python BootCamp")
        func()
        print("Today my last day of learning python basics in this course")
    return sub_welcome_method()
```

```python
def course_introduction() :
    print("This is an advance Python Course")

course_introduction()
```

```python
main_welcome(course_introduction)
```

```python
@main_welcome
def course_introduction() :
    print("This is an advance Python Course")
```

```python
## Decorators 

def my_decorator(func) :
    def wrapper() :
        print("Something is happening before the function is called")
        func()
        print("Something is happening after the function is called")
    return wrapper
```

```python
@my_decorator
def say_hello() :
    print("Hello")
```

```python
say_hello()
```

```python
## Decorators with arguments
def repeat(n) :
    def decorator(func) :
        def wrapper(*args, **kwargs) :
            for _ in range(n) :
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello() :
    print("Hello")
```

```python
say_hello()
```
