```python
def square(n) :
    for i in range(n):
        yield i ** 2
```

```python
square(4)
```

```python
for i in square(5) :
    print(i)
```

```python
a = square(5)
```

```python
def my_generator() :
    yield 1
    yield 2
    yield 3
```

```python
gen = my_generator()
```

```python
for val in gen :
    print(val)
```

```python
### Practical : Reading Large Files 
def read_large_file(filep_path):
    with open(filep_path, 'r') as file :
        for line in file :
            yield line
```

```python
file_path = 'large_file.txt'

for line in read_large_file(file_path) :
    print(line.strip())
```
