```python
my_list = [1,2,3,4,5]
for i in my_list :
    print(i)
```

```python
print(type(my_list))
```

```python
print(my_list)
```

```python
## Iterator
iterator = iter(my_list)
print(type(iterator))
```

```python
iterator
```

```python
## Iteratte through all the elements

next(iterator)
```

```python
new_list = [1,2,3,4,5]

iterator = iter(new_list)
```

```python
type(iterator)
```

```python
try :
    print(next(iterator))
except StopIteration :
    print("No more elements present in the iterator")
```

```python
## String Iterator 
my_string = "Hello world"
string_iterator = iter(my_string)
```

```python
print(next(string_iterator))
```
