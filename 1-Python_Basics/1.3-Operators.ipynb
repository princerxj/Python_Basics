```python
## Arithmetic Operation 
a = 10 
b = 5
add_result = a + b # Addition Operation 
sub_result = a - b #Subtraction Operation 
prod_result = a * b #Multiplication operation 
div_result = a / b # Division Operation 
floor_div_result = a//b # Floor Division Operation 
modulus_result = a % b # Modulus Operation 
exponential_result = a**b # Exponential 

print(add_result)
print(sub_result)
print(prod_result)
print(div_result)
print(floor_div_result)
print(modulus_result)
print(exponential_result)
```

```python
## Comparison Operators 
## Double equal to == , Equal To OPerator
a = 10 
b = 10
a == b
```

```python
str1 = "Krish"
str2 = "Krish"
str1 == str2
```

```python
## Not Equal to : !=
str1 != str2
```

```python
## Greater Than >
num1 = 45
num2 = 55 
num1 > num2
```

```python
## less than : <
num1 < num2
```

```python
## Greater than or equal to : >=
num1 >= num2
```
