```python
## Simple Calculator 
num1 = float(input("Enter the first Number : "))
num2 = float(input("Enter the second Number"))
sum = num1 + num2
diff = num1 - num2
mul = num1 * num2
quotient = num1/ num2
print("The sum is : ", sum)
print("The difference is : ", diff)
print("The product is : ", mul)
print("The quotient is : ", quotient)
```
