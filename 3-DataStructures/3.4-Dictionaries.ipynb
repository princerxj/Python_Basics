```python
student= {"name":"Krish", "age":32,"grade":24}
print(student)
print(type(student))
```

```python
## Always use Single Key
student= {"name":"Krish", "age":32,"name":24}
print(student)
```

```python
## Accessing Dictionary Elements
student= {"name":"Krish", "age":32,"grade":"A"}
print(student)
```

```python
print(student['grade'])
print(student['age'])

## Accessing using get() method
print(student.get('grade'))
print(student.get('name'))
print(student.get('age'))
print(student.get('last_name'))
print(student.get('last_name', "Not available"))
```

```python
## Modifying the dictionary Elements
## Dicionary are mutable , so we can add , update or delete elements
print(student)
```

```python
## Updating value for a key 
student["age"] = 33
print(student)
```

```python
## Adding a new key value pair
student["address"] = "India"
print(student)
```

```python
## Deleting a Key value pair
del student['grade']
print(student)
```

```python
## Dictionary Methods 
keys = student.keys() # get all the keys
print(keys)
values = student.values() # get all the values
print(values)

items = student.items() # get all the key value pairs
print(items)
```

```python
## Shallow Copy 
student_copy = student 
print(student)
print(student_copy)
```

```python
student["name"] = "Prince"
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy() #Shallow Copy
student["name"] = "Krish"
print(student)
print(student_copy1)
```

```python
## Iterating over Dictionaries
## We can use loops to iterate

## Iterating over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterating over the values
for values in student.values():
    print(values)
```

```python
## Iterate over key value pairs
for key, value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested Dictionaries
students = {
    "student1": {
        "name":"krish",
        "age" : 32
    },
    "student2":{
        "name":"Peter",
        "age":22
    }
}
print(students)
```

```python
### Access nested dictionary elements
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
## Iterating over nested dictionaries
for student_id, student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key, value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary Comprehension 
squares = {x:x**2 for x in range(5)}
print(squares)
```

```python
## Conditional Dictionary Comprehension 
even_squares = {x:x**2 for x in range(10) if x%2 == 0}
print(even_squares)
```

```python
### Practical Examples
## Use a dictionary to the frequency of elements in a list 
lst = [1,2,2,3,3,3,4,4,4,4]
freq = dict()
for num in lst:
    if num in freq:
        freq[num] = freq.get(num) + 1
    else :
        freq[num] = 1
print(freq) 
```

```python
## Merge two dictionaries into one
dict1 = {"a":1, "b":2}
dict2 = {"b":3, "c":4}
merged_dict = {**dict1, **dict2}
print(merged_dict)
```
